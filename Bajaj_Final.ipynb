```python
def is_valid_indian_number(phone_number):
  """
  Checks if a phone number is a valid Indian phone number.

  Args:
    phone_number: The phone number to check.

  Returns:
    True if the phone number is a valid Indian phone number, False otherwise.
  """
  if phone_number is None:
    return False
  # Remove any non-digit characters
  phone_number = ''.join(c for c in phone_number if c.isdigit())
  # Check if the phone number starts with +91 and has 10 digits
  if phone_number.startswith('+91') and len(phone_number) == 13:
    return True
  # Check if the phone number starts with 0 and has 10 digits
  elif phone_number.startswith('91') and len(phone_number) == 11:
    return True
  # Check if the phone number has 10 digits
  elif len(phone_number) == 10:
    return True
  else:
    return False

# Apply the function to the 'phoneNumber' column
df['isValidMobile'] = df['patientDetails'].apply(is_valid_indian_number)

# Display the updated DataFrame
print(df)
```
